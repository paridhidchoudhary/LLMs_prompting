```python
from transformers import pipeline
```

reference: https://huggingface.co/docs/transformers/main/en/tasks/prompting


```python
import torch
torch.manual_seed(0)

#Run inference with decoder-only models with the text-generation pipeline:
generator = pipeline('text-generation', model='openai-community/gpt2')
prompt = "Hello, I am a language model"
#wont work well for translation
#prompt= 'translate English to French: Hello, I am a language model'
generator(prompt, max_length=30)
```

```python
#encoder- decoder -> use text2text-generation
text2text_generator= pipeline('text2text-generation', model='google/flan-t5-base')
text2text_generator('translate English to French: Hello, I am a language model')
```
